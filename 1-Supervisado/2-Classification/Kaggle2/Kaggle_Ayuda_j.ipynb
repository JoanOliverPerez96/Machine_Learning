## Data processing

```python
def data_processing(train):
    train.set_index('Id', inplace=True)
    train["age"] = train["age"].str.replace("y", "").astype(int)
    train["age"] = train["age"].replace(0, 18).astype(int)
    train["MonthlyIncome"] = train["MonthlyIncome"].fillna(train["MonthlyIncome"].median())
    train["NumberOfDependents"] = train["NumberOfDependents"].fillna(0)
    return train
```

### 1. Definir X e y

```python
X = train.drop(["target"], axis=1)
y = train["target"]
### 2. Dividir X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
### 3. Asignar el modelo (vacío) a una variable

# Logistic Regression
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, y_train)
predict = logisticRegr.predict(X_test)
# Use score method to get accuracy of model
score = logisticRegr.score(X_test, y_test)
print("Logistic Regression Accuracy: ", score*100)
prediction = logisticRegr.predict(X_test)
df_pred = pd.DataFrame(prediction, columns=["target"])
predicions_proba = logisticRegr.predict_proba(X)
predicions_proba
name='Logistic Regression'
kfold = model_selection.KFold(n_splits=10) #Parte los datos en 10 trozos para usar validación cruzada / cross validation
cv_results = model_selection.cross_val_score(logisticRegr, X_train, y_train, cv=kfold, scoring='accuracy') # reg lineal -> Link docs
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
print(msg)
print("---------------------")


# Decission Tree
from sklearn.tree import DecisionTreeRegressor, plot_tree

dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
feat_imp = dtr.feature_importances_
feat_imp_df = pd.DataFrame({'features':X_train.columns, 'importances':feat_imp}).sort_values('importances', ascending=False)
feat_imp_df 
y_pred = dtr.predict(X_test)
print("DecisionTreeRegressor - no max depth")
print(y_pred)
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
import numpy as np

print("Coeficiente determinación", r2_score(y_test, y_pred))
print("MAE", mean_absolute_error(y_test, y_pred))
print("MAPE", mean_absolute_percentage_error(y_test, y_pred))
print("MSE", mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(mean_squared_error(y_test, y_pred)))
print("---------------------")

dtr_4 = DecisionTreeRegressor(max_depth=4)
dtr_4.fit(X_train,y_train)
feat_imp = dtr_4.feature_importances_
feat_imp_df = pd.DataFrame({'features':X_train.columns, 'importances':feat_imp}).sort_values('importances', ascending=False)
feat_imp_df 
y_pred = dtr_4.predict(X_test)
print("DecisionTreeRegressor - max depth = 4")
print(y_pred)
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
import numpy as np

print("Coeficiente determinación", r2_score(y_test, y_pred))
print("MAE", mean_absolute_error(y_test, y_pred))
print("MAPE", mean_absolute_percentage_error(y_test, y_pred))
print("MSE", mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(mean_squared_error(y_test, y_pred)))
```
